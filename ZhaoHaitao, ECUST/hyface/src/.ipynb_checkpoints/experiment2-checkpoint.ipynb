# 说明

为测试哪个波段对人脸识别的提升率最高，这里统一采用PCA和SVM方法，做了如下实验
1. 只采用灰度图像进行识别
2. 分别采用各频段与灰度合并组成特征进行识别

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time

from sklearn.pipeline import Pipeline 
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.preprocessing import StandardScaler

from sklearn.externals import joblib
from sklearn.metrics import accuracy_score, log_loss

from sklearn.decomposition import PCA
from sklearn.svm import SVC

import datasets

current_time = lambda: time.strftime('%Y-%m-%d %H:%M:%S',time.localtime())
saveresult = False
```

# Import Data

```python
X, y = datasets.load_X_y_with_gray('train', n_samples=800, dsize=(18, 22))
n_samples = X.shape[0]; n_channels = X.shape[3]

groupsplit = StratifiedShuffleSplit(n_splits=1, test_size=0.25, random_state=42)
```

```python
# 查看测试集样本中各类别的数目，训练集为其3倍
splitRes = groupsplit.split(X, y)
for train_index, test_index in splitRes:
    X_train, y_train = X[train_index], y[train_index]
    X_test,  y_test  = X[test_index],  y[test_index]
    
    print(np.bincount(y_train))
    print(np.bincount(y_test))
    datasets.show_cube(X_train[0], 5, True); datasets.show_cube(X_test[0], 5, True)
    plt.show()
```

# Define Model

```python
from sklearn.svm import SVC
# class sklearn.svm.SVC(C=1.0, kernel=’rbf’, degree=3, gamma=’auto’, coef0=0.0, shrinking=True,
#                       probability=False, tol=0.001, cache_size=200, 
#                       class_weight=None, verbose=False, max_iter=-1, 
#                       decision_function_shape=’ovr’, random_state=None)
model_type = 'svc'

steps = [
    ('standard', StandardScaler()),
    ('reduce_dim', PCA()), 
    ('classification', SVC(probability=True, kernel = 'linear')),
]
pipe = Pipeline(steps=steps)

hyper_params = {
    'reduce_dim__n_components': [int(18*22*(i/10)) for i in range(1, 10)],
    
    'classification__C': [0.2*i for i in range(1, 20)],
#     'classification__kernel': ['linear', 'rbf'],
    'classification__degree': [_ for _ in range(2, 5)],
}
searcher = RandomizedSearchCV(pipe, hyper_params, scoring='neg_log_loss')
# searcher = RandomizedSearchCV(pipe, hyper_params, scoring='accuracy')
```

# Train

```python
acc_train_saved = []; acc_test_saved = []
loss_train_saved = []; loss_test_saved = []

splitRes = groupsplit.split(X, y)
for train_index, test_index in splitRes:
    for c in range(n_channels):
        if c == 33:
            X_c = X[:,:,:,-1]
        else:
            X_c = np.dstack((X[:, :, :, c], X[:, :, :, -1]))
        
        X_train, y_train = X_c[train_index], y[train_index]
        X_test,  y_test  = X_c[test_index],  y[test_index]

        X_train = X_train.reshape((X_train.shape[0], -1))
        X_test  = X_test.reshape((X_test.shape[0], -1))
        
        searcher.fit(X_train, y_train)
        print(searcher.best_params_)
    #   print('The best score is: {:>.2f}'.format(searcher.best_score_))

        y_train_pred_proba = searcher.predict_proba(X_train)
        y_test_pred_proba = searcher.predict_proba(X_test)
        y_train_pred = searcher.predict(X_train)
        y_test_pred = searcher.predict(X_test)

        acc_train = accuracy_score(y_train_pred, y_train)
        acc_test  = accuracy_score(y_test_pred,  y_test )
        loss_train = log_loss(y_train, y_train_pred_proba) / 3. # 结果为加和，训练集样本数目为测试集3倍
        loss_test  = log_loss(y_test,  y_test_pred_proba)

        acc_train_saved.append(acc_train); acc_test_saved.append(acc_test)
        loss_train_saved.append(loss_train); loss_test_saved.append(loss_test)
        
        print_log = current_time() + ' channel {}/{} | acc_train: {:>.2%}, loss_train:{:>.6f} | acc_test: {:>.2%}, loss_test:{:>.6f}'.\
                        format(c, n_channels, acc_train, loss_train, acc_test, loss_test)
        print(print_log)

        if saveresult:
            if c == 33:
                model_name = '{}_gray_only.pkl'.format(model_type)
            else:
                model_name = '{}_gray_with_c{}.pkl'.format(model_type, c)
            save_path = os.path.join('F:/hyperspectral_face/models/exp2/{}'.format(model_type), model_name)
            joblib.dump(searcher.best_estimator_, save_path)
    
    print_log = '======================================================================================================='
    print(print_log)
```

```python
save_path = os.path.join('F:/hyperspectral_face/models/exp2', model_type)
if not os.path.exists(save_path): os.mkdir(save_path)

channels = np.arange(n_channels) + 1

plt.figure('accuracy', figsize=(30,10))
plt.title('signle channel with gray image'); plt.xlabel('channels'); plt.ylabel('accuracy/%')
plt.bar(channels[:33], np.array(acc_train_saved)[:33]*100, 
            alpha=0.9, width = 0.35, facecolor = 'lightskyblue', edgecolor = 'white', label='train')
plt.bar(channels[-1], np.array(acc_train_saved)[-1]*100, 
            alpha=0.9, width = 0.35, facecolor = 'lightblue', edgecolor = 'white', label='train')
plt.axhline(acc_train_saved[-1]*100)
plt.bar(channels[:33] + 0.35, np.array(acc_test_saved)[:33]*100, 
            alpha=0.9, width = 0.35, facecolor = 'yellowgreen', edgecolor = 'white', label='test')
plt.bar(channels[-1] + 0.35, np.array(acc_test_saved)[-1]*100, 
            alpha=0.9, width = 0.35, facecolor = 'lightgreen', edgecolor = 'white', label='test')
plt.axhline(acc_test_saved[-1]*100)
if saveresult:
    plt.savefig(os.path.join(save_path, 'accuracy.png'))

plt.figure('log_loss', figsize=(30,10))
plt.title('signle channel with gray image'); plt.xlabel('channels'); plt.ylabel('log_loss')
plt.bar(channels[:33], np.array(loss_train_saved)[:33], 
            alpha=0.9, width = 0.35, facecolor = 'lightskyblue', edgecolor = 'white', label='train')
plt.bar(channels[-1], np.array(loss_train_saved)[-1], 
            alpha=0.9, width = 0.35, facecolor = 'lightblue', edgecolor = 'white', label='train')
plt.axhline(loss_train_saved[-1])
plt.bar(channels[:33] + 0.35, np.array(loss_test_saved)[:33], 
            alpha=0.9, width = 0.35, facecolor = 'yellowgreen', edgecolor = 'white', label='test')
plt.bar(channels[-1] + 0.35, np.array(loss_test_saved)[-1], 
            alpha=0.9, width = 0.35, facecolor = 'lightgreen', edgecolor = 'white', label='test')
plt.axhline(loss_test_saved[-1])
if saveresult:
    plt.savefig(os.path.join(save_path, 'log_loss.png'))

# plt.legend(loc="upper left")
plt.show()
```

```python
# 载入测试数据
X_, y_ = datasets.load_X_y_with_gray('test', n_samples=-1, dsize=(18, 22))
idx = (y_<=10); X_ = X_[idx]; y_ = y_[idx]

for c in range(n_channels):
    model_path = os.path.join('F:/hyperspectral_face/models/exp2', model_type)
    if c == 33:
        model_name = '{}_gray_only.pkl'.format(model_type)
        X__ = X_[:, :, :, -1]
    else:
        model_name = '{}_gray_with_c{}.pkl'.format(model_type, c)
        X__ = np.dstack((X_[:, :, :, c], X_[:, :, :, -1]))
    
    model_name = os.path.join(model_path, model_name)
    model = joblib.load(model_name)
    
    X__ = X__.reshape((X__.shape[0], -1))
    
    y_pred = model.predict(X__)

    print('pred: ', y_pred)
    print('true: ', y_)
    
    acc  = accuracy_score(y_pred,  y_ )
    print_log = 'channel: {:>2}/{:>2} | acc: {:>.2%}'.\
                        format(c, n_channels, acc)
    print(print_log)
    print('===============================')
```
