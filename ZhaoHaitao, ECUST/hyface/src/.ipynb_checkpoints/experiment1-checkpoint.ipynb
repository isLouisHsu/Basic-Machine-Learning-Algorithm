# 说明

1. 仅采用正脸图
1. 考虑计算量和数据的缺失，只使用10个人的人脸数据，即10分类问题
2. 将原有数据集扩增到每个人脸80个样本
2. 分别用SVM与RandomForest做了如下实验
    1. 合并通道训练
    2. 分离通道训练，统计各通道训练集和测试集上准确率与损失
3. 损失采用交叉熵
4. 通过随机搜索确定超参数


* SVM方法较好
* 从损失上看，测试集损失为训练集10倍，过拟合？
* 从准确率看，测试集与训练集差不多
* 5至31通道准确率与损失差不多

````python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time

from sklearn.pipeline import Pipeline 
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.preprocessing import StandardScaler

from sklearn.externals import joblib
from sklearn.metrics import accuracy_score, log_loss

from sklearn.decomposition import PCA

import datasets

current_time = lambda: time.strftime('%Y-%m-%d %H:%M:%S',time.localtime())
saveresult = True
````

# Import Data

````python
X, y = datasets.load_X_y('SampleImages_F_augmented', mode='train', n_samples=800, dsize=(18, 22))
n_samples = X.shape[0]; n_channels = X.shape[3]

groupsplit = StratifiedShuffleSplit(n_splits=1, test_size=0.25, random_state=42)
````

````python
# 查看测试集样本中各类别的数目，训练集为其3倍
splitRes = groupsplit.split(X, y)
for train_index, test_index in splitRes:
    X_train, y_train = X[train_index], y[train_index]
    X_test,  y_test  = X[test_index],  y[test_index]
    
    print(np.bincount(y_train))
    print(np.bincount(y_test))
    datasets.show_cube(X_train[0]); datasets.show_cube(X_test[0])
    plt.show()
````

# Define Models

## sklearn.svm.SVC

````python
from sklearn.svm import SVC
# class sklearn.svm.SVC(C=1.0, kernel=’rbf’, degree=3, gamma=’auto’, coef0=0.0, shrinking=True,
#                       probability=False, tol=0.001, cache_size=200, 
#                       class_weight=None, verbose=False, max_iter=-1, 
#                       decision_function_shape=’ovr’, random_state=None)
model_type = 'svc'

steps = [
    ('standard', StandardScaler()),
    ('reduce_dim', PCA()), 
    ('classification', SVC(probability=True, kernel = 'linear')),
]
pipe = Pipeline(steps=steps)

hyper_params = {
    'reduce_dim__n_components': [int(18*22*(i/10)) for i in range(1, 10)],
    
    'classification__C': [0.2*i for i in range(1, 20)],
#     'classification__kernel': ['linear', 'rbf'],
    'classification__degree': [_ for _ in range(2, 5)],
}
searcher = RandomizedSearchCV(pipe, hyper_params, scoring='neg_log_loss')
# searcher = RandomizedSearchCV(pipe, hyper_params, scoring='accuracy')
````

## sklearn.ensemble.RandomForestClassifier

````python
from sklearn.ensemble import RandomForestClassifier
# class sklearn.ensemble.RandomForestClassifier(n_estimators=10, criterion=’gini’, max_depth=None,
#                                             min_samples_split=2, min_samples_leaf=1,
#                                             min_weight_fraction_leaf=0.0,
#                                             max_features=’auto’,
#                                             max_leaf_nodes=None,
#                                             min_impurity_decrease=0.0,
#                                             min_impurity_split=None, bootstrap=True, oob_score=False, n_jobs=1,
#                                             random_state=None, verbose=0,
#                                             warm_start=False, class_weight=None)
model_type = 'forest'

steps = [
    ('standard', StandardScaler()),
    ('reduce_dim', PCA()), 
    ('classification', RandomForestClassifier(n_estimators=20)),
]
pipe = Pipeline(steps=steps)

hyper_params = {
    'reduce_dim__n_components': [int(18*22*(i/10)) for i in range(1, 10)],
    'reduce_dim__whiten': [False, True],
    
    'classification__max_depth': [2*i for i in range(1, 10)],
    'classification__min_samples_split': [2*i for i in range(1, 4)],
    'classification__min_samples_leaf': [i for i in range(1, 10)],
}
searcher = RandomizedSearchCV(pipe, hyper_params, scoring='neg_log_loss')
# searcher = RandomizedSearchCV(pipe, hyper_params, scoring='accuracy')
````

# Training & Searching

## Split

````python
acc_train_saved = []; acc_test_saved = []
loss_train_saved = []; loss_test_saved = []

for c in range(n_channels):
    X_channel_c = X[:, :, :, c]
    
    splitRes = groupsplit.split(X_channel_c, y)
    for train_index, test_index in splitRes:
        X_train, y_train = X_channel_c[train_index], y[train_index]
        X_test,  y_test  = X_channel_c[test_index],  y[test_index]
        
        X_train = X_train.reshape((X_train.shape[0], -1))
        X_test  = X_test.reshape((X_test.shape[0], -1))
        
        searcher.fit(X_train, y_train)
        print(searcher.best_params_)
#         print('The best score is: {:>.2f}'.format(searcher.best_score_))

        y_train_pred_proba = searcher.predict_proba(X_train)
        y_test_pred_proba = searcher.predict_proba(X_test)
        y_train_pred = searcher.predict(X_train)
        y_test_pred = searcher.predict(X_test)

        acc_train = accuracy_score(y_train_pred, y_train)
        acc_test  = accuracy_score(y_test_pred,  y_test )
        loss_train = log_loss(y_train, y_train_pred_proba) / 3. # 结果为加和，训练集样本数目为测试集3倍
        loss_test  = log_loss(y_test,  y_test_pred_proba)
        
        acc_train_saved.append(acc_train); acc_test_saved.append(acc_test)
        loss_train_saved.append(loss_train); loss_test_saved.append(loss_test)
        
        print_log = current_time() + ' channel: {:>2}/{:>2} | acc_train: {:>.2%}, loss_train:{:>.6f} \
                                                            | acc_test: {:>.2%}, loss_test:{:>.6f}'.\
                        format(c, n_channels, acc_train, loss_train, acc_test, loss_test)
        print(print_log)
        
        if saveresult:
            model_name = 'searcher_{}_channel_{:d}.pkl'.format(model_type, c)
            save_path = os.path.join('F:/hyperspectral_face/models/split/{}'.format(model_type), model_name)
            joblib.dump(searcher.best_estimator_, save_path)
    
    print_log = '======================================================================================================='
    print(print_log)
````

````python
save_path = os.path.join('F:/hyperspectral_face/models/split', model_type)
if not os.path.exists(save_path): os.mkdir(save_path)

channels = np.arange(33) + 1

plt.figure('accuracy', figsize=(30,10))
plt.title('signal channel'); plt.xlabel('channels'); plt.ylabel('accuracy/%')
plt.bar(channels, np.array(acc_train_saved)*100, 
            alpha=0.9, width = 0.35, facecolor = 'lightskyblue', edgecolor = 'white', label='train')
plt.bar(channels + 0.35, np.array(acc_test_saved)*100, 
            alpha=0.9, width = 0.35, facecolor = 'yellowgreen', edgecolor = 'white', label='test')
if saveresult:
    plt.savefig(os.path.join(save_path, 'accuracy.png'))

plt.figure('log_loss', figsize=(30,10))
plt.title('signal channel');plt.xlabel('channels'); plt.ylabel('log_loss')
plt.bar(channels, np.array(loss_train_saved), 
            alpha=0.9, width = 0.35, facecolor = 'lightskyblue', edgecolor = 'white', label='train')
plt.bar(channels + 0.35, np.array(loss_test_saved), 
            alpha=0.9, width = 0.35, facecolor = 'yellowgreen', edgecolor = 'white', label='test')
if saveresult:
    plt.savefig(os.path.join(save_path, 'log_loss.png'))

# plt.legend(loc="upper left")
plt.show()
````

## Merge

````python
splitRes = groupsplit.split(X, y)
for train_index, test_index in splitRes:
    X_train, y_train = X[train_index], y[train_index]
    X_test,  y_test  = X[test_index],  y[test_index]
        
    X_train = X_train.reshape((X_train.shape[0], -1))
    X_test  = X_test.reshape((X_test.shape[0], -1))
        
    searcher.fit(X_train, y_train)
    print(searcher.best_params_)
#   print('The best score is: {:>.2f}'.format(searcher.best_score_))

    y_train_pred_proba = searcher.predict_proba(X_train)
    y_test_pred_proba = searcher.predict_proba(X_test)
    y_train_pred = searcher.predict(X_train)
    y_test_pred = searcher.predict(X_test)

    acc_train = accuracy_score(y_train_pred, y_train)
    acc_test  = accuracy_score(y_test_pred,  y_test )
    loss_train = log_loss(y_train, y_train_pred_proba) / 3. # 结果为加和，训练集样本数目为测试集3倍
    loss_test  = log_loss(y_test,  y_test_pred_proba)
        
    print_log = current_time() + '  acc_train: {:>.2%}, loss_train:{:>.6f} | acc_test: {:>.2%}, loss_test:{:>.6f}'.\
                    format(acc_train, loss_train, acc_test, loss_test)
    print(print_log)
        
    if saveresult:
        model_name = '{}.pkl'.format(model_type)
        save_path = os.path.join('F:/hyperspectral_face/models/merge/{}'.format(model_type), model_name)
        joblib.dump(searcher.best_estimator_, save_path)
    
print_log = '======================================================================================================='
print(print_log)
````
